```python
import pandas as pd
```

```python
data = pd.read_csv("loan_approval_dataset.csv")
```

```python
data
```

```python
data = data.drop(columns=["loan_id"])
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data.columns.tolist())
```

```python
data = data.rename(columns={"loan_status": "loan_approved"})

data["loan_approved"] = (
    data["loan_approved"]
    .astype(str)
    .str.strip()
    .map({
        "Approved": True,
        "Rejected": False
    })
)
```

```python
data = data.rename(columns={"education": "graduated"})

data["graduated"] = (
    data["graduated"]
    .astype(str)
    .str.strip()
    .map({
        "Graduate": True,
        "Not Graduate": False
    })
)
```

```python
data["self_employed"] = (
    data["self_employed"]
    .astype(str)
    .str.strip()
    .map({
        "Yes": True,
        "No": False
    })
)
```

```python
data
```

```python
print(data.isna().sum())
```

```python
data.to_csv("cleaned_loan_data.csv", index=False)
```
